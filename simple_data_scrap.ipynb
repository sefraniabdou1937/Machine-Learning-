```python
import pandas as pd
import re

# Fonction pour scraper les données d'une saison
def scrape_season_data(start_year, end_year):
    # Construire l'URL pour la saison donnée
    season = f"{start_year}-{end_year}"
    base_url_schedule = f'https://fbref.com/en/squads/1df6b87e/{season}/matchlogs/all_comps/schedule/Sheffield-United-Scores-and-Fixtures-All-Competitions'
    base_url_shooting = f'https://fbref.com/en/squads/1df6b87e/{season}/matchlogs/all_comps/shooting/Sheffield-United-Match-Logs-All-Competitions'

    # Charger les données
    try:
        df = pd.read_html(base_url_schedule, attrs={"id": "matchlogs_for"})[0]
        df2 = pd.read_html(base_url_shooting, attrs={"id": "matchlogs_for"})[0]
    except Exception as e:
        print(f"Erreur lors du scraping pour la saison {season}: {e}")
        return None

    # Filtrer uniquement les matches de Premier League
    filtered_df1 = df[df['Comp'] == 'Premier League']
    df2.columns = df2.columns.droplevel()  # Supprimer les niveaux multiples dans les colonnes
    filtered_df2 = df2[df2['Comp'] == 'Premier League']

    # Fusionner les deux DataFrames
    merged_data = pd.concat([filtered_df1.reset_index(drop=True), filtered_df2.reset_index(drop=True)], axis=1)
    merged_data = merged_data.loc[:, ~merged_data.columns.duplicated()]  # Supprimer les colonnes dupliquées

    # Extraire l'année et le nom de l'équipe
    match = re.search(r'/squads/\w+/(\d{4}-\d{4})/.*?/([^/]+)-Scores-and-Fixtures', base_url_schedule)
    if match:
        year = match.group(1)  # Année
        team_name = match.group(2).replace('-', ' ')  # Nom de l'équipe

    # Enregistrer les données dans un fichier CSV
    filename = f'{year}_{team_name}.csv'
    merged_data.to_csv(filename, index=False, encoding='utf-8')
    print(f"Fichier sauvegardé: {filename}")

# Scraper les données pour les 5 dernières saisons
for i in range(5):
    start_year = 2023 - i
    end_year = 2024 - i
    scrape_season_data(start_year, end_year)
```
